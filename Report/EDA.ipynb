# Read the dataset

```python
images=[]
labels=[]
dimension = []
```

```python
#This function return the class name according to the folder
def assign_label(img,flower_type):
    return flower_type
```

```python
#Function to read all images in the dataset
#The dimension array used for storing dimension. The lable array is for the class name
def make_train_data(flower_type):
    DIR = '/' + flower_type + '/'
    current_dir = os.curdir + "/Flowers" + DIR

    for img in tqdm(os.listdir(current_dir)):
        label=assign_label(img,flower_type)    
        path = os.path.join(current_dir,img)
        img = cv2.imread(path,cv2.IMREAD_COLOR)
        images.append(img)
        labels.append(str(label))
        dimension.append(img.shape)
        
```

```python
#Read folder with corresponding name and read images, save to array data
classes = ['Tana', 'Rosy', 'Pingpong','Lisianthus', 'Hydrangeas', 'Chrysanthemum', 'Calimerio', 'Babi']

for item in classes:
    make_train_data(item)
```
